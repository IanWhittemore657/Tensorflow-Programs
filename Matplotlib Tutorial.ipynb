# Line Chart 

```
import numpy as np
import matplotlib.pyplot as plt
```

```
#this creates a 1-D array with 1000 even points between 0 and 20

x = np.linspace(0,20,1000)
```

```
#this is some fake data for the y axis
y = np.sin(x) + 0.2 * x
```

```
#this will plot the data into a chart
plt.plot(x,y)


#if running the code not on a notebook then will need to use plt.show
```

```
#label the x axis and y axis

plt.plot(x,y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("My plot");
#the semicolon removes the output and only shows the chart 
```

# **Scatterplot**

```
#create random data

X = np.random.randn(100,2)
plt.scatter(X[:,0],X[:,1])

```

```
#scatterplot with different colours for different classes

X = np.random.randn(200,2)

X[:50] += 3

#generate an array of 0s
Y = np.zeros(200)

#all the points centered at 3 3 have a label 

Y[:50] = 1

plt.scatter(X[:,0],X[:,1],c=Y)
```

# **Histogram**

```
X = np.random.randn(10000)
```

```
plt.hist(X);
```

```
#using bins makes more bars 
plt.hist(X,bins=50);
```

# **Plotting Images**

```
#grab a url 

#then use pillow to get image

from PIL import Image

im = Image.open("image.png")

#to see what type just use 

type(im)

#Then to make it an array in numpy 

arr = np.array(im)

#then arr is the image in an array

#to see the size of the image use arr.shape

arr.shape

#this will give an output such as (512,512,3) which the first two are the pixels and the last three is the rgb colour

#then to show the image via the array

plt.imshow(arr);

#or use the actual image
plt.imshow(im)


#to collapse the rgb 

gray = arr.mean(axis = 2)

gray.shape
plt.imshow(gray)

#then to make it greyscale 

plt.imshow(gray,cmap="gray")
```
