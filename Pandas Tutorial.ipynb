# **The apply() function**

```
def date_to_year(row):
  return int(row["date"].split('-')[0])

df.apply(date_to_year,axis =1)
```

```
#we can assign this to a new column 

df["year"] = df.apply(date_to_year,axis =1)

df.head()
```
