# **The Dot Product**

```
#one dimensional arrays

a = np.array([1,2])
b = np.array([3,4])

```

```
#multiple the two element arrays together 

dot = 0
for e,f in zip(a,b):
  dot += e * f

dot
```

```
dot = 0
for i in range(len(a)):
  dot += a[i] * b[i]
dot
```

```
#this give element wise multiplication similar to above but not adding them togther
a * b
```

```
#this gives the sum though of a + b 
np.sum(a * b)

#this gives the same by using this though 
(a * b).sum()
```

```
#another way can be as simple as though using 

np.dot(a,b)

#or even using this gives the same answer
a.dot(b)
```

```
#in newer versions of numpy it can be as simple to do this by using just the @ symbol
a @ b
```

```
#maginute of the vector is :
amag = np.sqrt((a * a).sum())
amag
```

```
#linear algebra tools 

np.linalg.norm(a)
```

```
#now that we know the above we can work out the cosine of the angle
cosangle = a.dot(b) /(np.linalg.norm(a)* np.linalg.norm(b))

cosangle
```

```
#if we want the actual angle then we use arccros
angle = np.arccos(cosangle)
angle
```

# **Solving Linear Systems:**

Example

Admin fee is 1.50 for children and 4.00 for adults , on a certain day there was 2200 people and it collected 5050 worth of money.

How many children and how many adults were there?

2 Equations 
2 Unknonws 

x1 + x2 = 2200
1.5x1 + 4x1 = 5050

```
# we can convert this into a matrix 

#so A = (1,1),(1.5,4) and B = (2200),(5050)
```

```
A = np.array([[1,1],[1.5,4]])
B = np.array([2200,5050])
```

```
np.linalg.solve(A,B)
```

```
#therefore there are 1500 adults and 500 children
```
